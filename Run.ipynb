```python
import threading
def main():
    start_time = time.ctime()
    print("Run Main_Function：{}".format(start_time))
    
    t1 = threading.Thread(target=Th_one)
    t1.setName("TH-1")
    t1.start()

    t2 = threading.Thread(target=Th_two)
    t2.setName("TH-2")
    t2.start()

    t3 = threading.Thread(target=Th_three)
    t3.setName("TH-3")
    t3.start()
    
    t4 = threading.Thread(target=Th_four)
    t4.setName("TH-4")
    t4.start()

    # enumerate Get the running process
    for i in threading.enumerate():
        # get(thread)Name 
        name = i.getName()
        print("正在運行的程式：{}".format(name))
    # get the running (thread) number
    num = threading.activeCount()
    print("正在運行的程式數量{}个".format(num))
    print("Finish \n")
    over_time = time.ctime()
    
    print("Start Time：{}".format(start_time))
    print("Over Time：{}".format(over_time))
if __name__ == '__main__':
    main()
```
